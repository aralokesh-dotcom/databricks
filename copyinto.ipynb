```python
# Step 1: Convert JSON to Parquet
json_path = "/Volumes/main/default/raw_data_volume/exported_users_json/"
parquet_path = "/Volumes/main/default/raw_data_volume/parquet_files_1/"

# Read JSON files
df_json = spark.read.format("json").load(json_path)

# Write as Parquet
df_json.write.mode("overwrite").format("parquet").save(parquet_path)

print(f"Converted JSON files from {json_path} to Parquet at {parquet_path}")
print(f"Schema: {df_json.schema}")
```

```python
%sql
CREATE TABLE IF NOT EXISTS main.default.my_delta_json USING DELTA;
use main.default;
COPY INTO my_delta_json
FROM '/Volumes/main/default/raw_data_volume/parquet_files_1/'
FILEFORMAT = PARQUET
COPY_OPTIONS ('mergeSchema' = 'true');

SELECT * FROM  main.default.my_delta_json LIMIT 10;
```
